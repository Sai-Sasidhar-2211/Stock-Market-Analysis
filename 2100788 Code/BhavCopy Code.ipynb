```python
#Importing the required libraries

import os
import pandas as pd
```

```python
# snippet to combine all the different tables into one single file

master_df = pd.DataFrame()
for file in os.listdir(os.getcwd()):
    if file.endswith('.csv'):
        master_df = master_df.append(pd.read_csv(file))       
master_df.to_csv("Final_BhavCopy.csv", index=False)
```

```python
#Converting date column to datetime dtype and setting an index for the converted column

df['DATE'] = pd.to_datetime(df['TIMESTAMP'])
df.set_index(['TIMESTAMP'], inplace=True)
df.head()
```

```python
df.info()
```

```python
#Copying the Selected Companies into a seperate variable along with the code snippet to copy the data 

copy = ["TCS","NIFTYBEES","RELIANCE","NIF100BEES","SBIN","INFY"]
copy_df = pd.DataFrame()
for i in range(len(copy)):
    copy_df = copy_df.append(df.loc[(df.SYMBOL == copy[i])])
```

```python
copy_df.head()
```

```python
#Checking if the selected companies are safely copied into the seperate table along with thesize of the table created

print(copy_df.SYMBOL.unique())
print(copy_df.shape)
```

```python
#Copying the Selected Companies into a seperate variable

share = ["SYMBOL","OPEN","HIGH", "LOW", "CLOSE", "TOTTRDQTY"]
share_df = df[share]
share_df.head()
```

```python
#Creating a correlation between the various companies from the seperated list

corr = [['TCS','INFY'],['TCS','NIFTYBEES'],['TCS','NIF100BEES'],['TCS','RELIANCE'],['RELIANCE','SBIN']]
for i in range(len(corr)):
    print(corr[i][0]," - ",corr[i][1])
    print(copy_df.loc[(copy_df.SYMBOL == corr[i][0])].corrwith(copy_df.loc[(copy_df.SYMBOL == corr[i][1])],axis = 1))
```

```python
#Adding the 2-Day Average and 5-Day Average as a column into the new sub-table created 

share_df['2D_Avg']= share_df['CLOSE'].rolling(2).mean()
share_df['5D_Avg'] = share_df['CLOSE'].rolling(5).mean()
```

```python
#Checking if the data is containing any null values

share_df.isnull().any()
```

```python
share_df.head()
```

```python
#Replacing NaN data to some understandable type such as a '-' 

share_df['2D_Avg'].fillna('-',inplace = True)
share_df['5D_Avg'].fillna('-',inplace = True)
```
